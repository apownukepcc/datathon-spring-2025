```python
# Google Colab Notebook: Merge Two Data Files Using Pandas

# Import necessary libraries
import pandas as pd

# Create two sample dataframes (simulating data files)
data1 = {'ID': [1, 2, 3, 4], 'Name': ['Alice', 'Bob', 'Charlie', 'David'], 'Score': [85, 90, 88, 76]}
data2 = {'ID': [3, 4, 5, 6], 'Name': ['Charlie', 'David', 'Eve', 'Frank'], 'Grade': ['B', 'C', 'A', 'B']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Display the original dataframes
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Perform different types of merges using Name column only
inner_merge = pd.merge(df1, df2, on='Name', how='inner')
left_merge = pd.merge(df1, df2, on='Name', how='left')
right_merge = pd.merge(df1, df2, on='Name', how='right')
outer_merge = pd.merge(df1, df2, on='Name', how='outer')

# Display results
print("\nInner Merge (Only matching rows):")
print(inner_merge)

print("\nLeft Merge (All from df1, matching from df2):")
print(left_merge)

print("\nRight Merge (All from df2, matching from df1):")
print(right_merge)

print("\nOuter Merge (All data, merging where possible):")
print(outer_merge)
```
